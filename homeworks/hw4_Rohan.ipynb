```python
import requests
from collections import defaultdict
from pandas import DataFrame
url = "https://raw.github.com/rdhyee/working-open-data/5ef3932b4ff7cadf1f06ca01eb852ad71361894a/data/census/DataDict.txt"

r = requests.get(url)

#print r,r.status_code,r.headers
#print "\n"
#print type(r.content)

datalist = r.content.split("\n")

numrows = len(datalist)
ditem = list()

# used fixed length substring as file has a fixed length format

for i in range(1,numrows-1):
	#print datalist[i]
	ditem.append(('Data_Item',datalist[i][:9]))
	ditem.append(('Item_Description',datalist[i][10:114]))
	#print datalist[i][10:114]
	ditem.append(('Unit',datalist[i][115:120]))
	#print datalist[i][115:120]
	ditem.append(('Decimal',datalist[i][121:128]))
	#print datalist[i][121:128]
	ditem.append(('US_Total',datalist[i][129:141]))
	#print datalist[i][129:141]

	ditem.append(('Minimum',datalist[i][142:152]))
	#print datalist[i][142:152]

	ditem.append(('Maximum',datalist[i][153:163]))
	#print datalist[i][153:163]
	ditem.append(('Source',datalist[i][164:170]))
	#print datalist[i][164:170]
	#print "\n"

#print ditem

d = defaultdict(list)

for k,v in ditem:
	d[k].append(v)

#print d


dframe = DataFrame(d)

#print dframe


```
